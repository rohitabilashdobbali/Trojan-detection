```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Load the labeled dataset
data = pd.read_csv('final_dataset.csv')

print(data.head())
```

```python

label_encoder = LabelEncoder()
data['Type'] = label_encoder.fit_transform(data['Type'])


X = data[['Current', 'Volt', 'Power']]
y = data['Type']


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Splitting the training and testing data  (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
print(f'Training Data: {X_train.shape}, Test Data: {X_test.shape}')
```

```python
import pandas as pd

# Convert X_test and y_test to DataFrames or Series before saving
X_test_df = pd.DataFrame(X_test, columns=X.columns)  # Make sure to use original column names
y_test_df = pd.Series(y_test, name='label')  # Give the label column a name, e.g., 'label'

# Save as CSV files
X_test_df.to_csv('X_test.csv', index=False)
y_test_df.to_csv('y_test.csv', index=False)
```

```python
# Initialize classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier 
rf_classifier.fit(X_train, y_train)
```

```python
# Prediction for testing data
y_pred = rf_classifier.predict(X_test)
```

```python
#  accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')


print('\nClassification Report:')
print(classification_report(y_test, y_pred, target_names=['NO Trojan', 'Trojan']))

# confusion matrix
print('\nConfusion Matrix:')
print(confusion_matrix(y_test, y_pred))
```

```python
import joblib

# Save the trained model
joblib.dump(rf_classifier, 'random_forest_trojan_model.pkl')

# To load the model later:
# rf_classifier = joblib.load('random_forest_trojan_model.pkl')
```

```python

unlabeled_data = pd.read_csv('unlabeled_dataset.csv')


X_new = scaler.transform(unlabeled_data[['Current', 'Volt', 'Power']])


y_new_pred = rf_classifier.predict(X_new)


predictions = label_encoder.inverse_transform(y_new_pred)


print(predictions)
```

```python
#  Save the predictions to a CSV file
unlabeled_data['Predicted_Type'] = predictions
unlabeled_data.to_csv('predicted_unlabeled_dataset.csv', index=False)
print('Predictions saved to predicted_unlabeled_dataset.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['NO Trojan', 'Trojan'], yticklabels=['NO Trojan', 'Trojan'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
